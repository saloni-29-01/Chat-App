## Q1 — Call Gemini 2.5 Flash via the OpenAI-compatible API

Use the `openai` Python package pointed at Google's OpenAI-compatible endpoint to make a single chat completion call to **gemini-2.5-flash**.

**Hints:**
- Set `base_url` to `https://generativelanguage.googleapis.com/v1beta/openai/`
- Set `api_key` to your Google AI Studio API key
- Use `client.chat.completions.create(...)` as you normally would with OpenAI

```python
# Install OpenAI package
!pip install openai --quiet

# Imports
from openai import OpenAI
import getpass

# Enter your Google AI Studio API key (NOT OpenAI key)
api_key = getpass.getpass("Enter your Google AI Studio API key: ")

# Create client with Gemini endpoint
client = OpenAI(
    api_key=api_key,
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/"
)

# Make chat completion request
response = client.chat.completions.create(
    model="gemini-2.5-flash",
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "Explain Artificial Intelligence in simple terms."}
    ]
)

# Print output
print(response.choices[0].message.content)
```

## Q3 — Build a Gradio Interface for Text Summarisation

Create a **Gradio** web app that takes a block of text as input and returns a concise summary. Use any LLM of your choice to generate the summary.

**Hints:**
- Define a `summarise(text)` function that calls an LLM and returns the summary string
- Wrap it with `gr.Interface(fn=..., inputs=gr.Textbox(...), outputs=gr.Textbox(...))`
- Call `.launch()` to start the app — use `share=True` to get a public link in Colab

```python
# Install libraries
!pip install gradio litellm --quiet

# Imports
import gradio as gr
import litellm
import os
import getpass

# Enter Gemini API key
os.environ["GEMINI_API_KEY"] = getpass.getpass("Enter your Gemini API key: ")

# ---------------------------
# Summarisation Function
# ---------------------------
def summarise(text):
    if not text.strip():
        return "Please enter some text."

    response = litellm.completion(
        model="gemini/gemini-2.5-flash",
        messages=[
            {"role": "system", "content": "Summarise the following text concisely."},
            {"role": "user", "content": text}
        ]
    )

    return response['choices'][0]['message']['content']

# ---------------------------
# Gradio Interface
# ---------------------------
interface = gr.Interface(
    fn=summarise,
    inputs=gr.Textbox(lines=10, placeholder="Paste your text here..."),
    outputs=gr.Textbox(label="Summary"),
    title="Text Summarisation App",
    description="Enter a block of text and get a concise summary."
)

# Launch app
interface.launch(share=True)
```

## Q4 — Build a Crew with CrewAI

Use **CrewAI** to create a small multi-agent crew. Define at least two agents with distinct roles and one task per agent, then kick off the crew.

**Hints:**
- Import `Agent`, `Task`, and `Crew` from `crewai`
- Each `Agent` needs a `role`, `goal`, and `backstory`
- Each `Task` needs a `description`, `expected_output`, and an assigned `agent`
- Create a `Crew(agents=[...], tasks=[...])` and call `.kickoff()`

```python
# Install required libraries
!pip install crewai litellm langchain-community --quiet

# Imports
from crewai import Agent, Task, Crew
import os
import getpass
from langchain_community.chat_models import ChatLiteLLM # Import ChatLiteLLM

# Set Gemini API key
os.environ["GEMINI_API_KEY"] = getpass.getpass("Enter your Gemini API key: ")

# Initialize LiteLLM for Gemini
gemini_llm = ChatLiteLLM(model="gemini/gemini-2.5-flash")

# ---------------------------
# Define Agents
# ---------------------------

research_agent = Agent(
    role="Research Analyst",
    goal="Collect key information about a topic",
    backstory="An expert researcher skilled at finding and summarizing important insights.",
    verbose=True,
    llm=gemini_llm # Assign the LiteLLM instance here
)

writer_agent = Agent(
    role="Content Writer",
    goal="Transform research into clear and simple content",
    backstory="A professional writer who explains complex ideas in an easy way.",
    verbose=True,
    llm=gemini_llm # Assign the LiteLLM instance here
)

# ---------------------------
# Define Tasks
# ---------------------------

research_task = Task(
    description="Research the benefits of Artificial Intelligence in education and list 5 key points.",
    expected_output="5 bullet points explaining AI benefits in education.",
    agent=research_agent
)

writing_task = Task(
    description="Use the research points to write a short paragraph summary.",
    expected_output="A clear and concise paragraph about AI in education.",
    agent=writer_agent
)

# ---------------------------
# Create Crew
# ---------------------------

crew = Crew(
    agents=[research_agent, writer_agent],
    tasks=[research_task, writing_task],
    verbose=True
)

# ---------------------------
# Execute Crew
# ---------------------------

result = crew.kickoff()

# Print result
print("\nFinal Output:\n")
print(result)
```

## Q5 — Build a Chain with LangChain

Use **LangChain** to build a simple prompt → LLM → output chain using LCEL (LangChain Expression Language).

**Hints:**
- Import `ChatPromptTemplate` from `langchain_core.prompts`
- Import a chat model, e.g. `ChatGoogleGenerativeAI` from `langchain_google_genai`
- Import `StrOutputParser` from `langchain_core.output_parsers`
- Chain them together with `|`: `prompt | llm | parser`
- Invoke the chain with `.invoke({"input": "..."})`

```python
# Install required libraries
!pip install langchain langchain-google-genai --quiet

# Imports
from langchain_core.prompts import ChatPromptTemplate
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.output_parsers import StrOutputParser
import os
import getpass

# Set Gemini API key
os.environ["GOOGLE_API_KEY"] = getpass.getpass("Enter your Gemini API key: ")

# ---------------------------
# Create Prompt Template
# ---------------------------
prompt = ChatPromptTemplate.from_template(
    "Summarise the following text in a concise way:\n\n{input}"
)

# ---------------------------
# Create LLM
# ---------------------------
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash"
)

# ---------------------------
# Output Parser
# ---------------------------
parser = StrOutputParser()

# ---------------------------
# Build Chain (LCEL)
# ---------------------------
chain = prompt | llm | parser

# ---------------------------
# Run Chain
# ---------------------------
result = chain.invoke({
    "input": "Artificial Intelligence is transforming industries by automating tasks, improving efficiency, and enabling smarter decision-making."
})

# Print result
print("Summary:\n")
print(result)
```
